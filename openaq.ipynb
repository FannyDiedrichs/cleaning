```python
missing_values = ["n/a", "na", "--"]

df = pd.read_csv("openaq.csv", sep=";", na_values = missing_values)
df.columns = ['Country_code', 'City', 'Location', 'Coordinates', 'Pollutant', 'Source_Name', 'Unit', 'Value', 'Last_Updated', 'Country_Label']
df
```

## Supprimer les données qui ne concernent pas la journée du 06-08-2021.

```python
df['Last_Updated'] = pd.to_datetime(df['Last_Updated'], utc=True)
df['Date'] = df['Last_Updated'].dt.date.astype(str)
date_dfa = df[~(df['Date'] < '2021-08-06')]
date_df = date_dfa[~(df['Date'] > '2021-08-06')]
date_df.to_csv('openaq_date.csv') 
```
